## 5.1.2 connection 창  
---

- [MySQL Connections]창    
접속될 서버와 사용자, 포트를 선택한 후 접속 시도  
MySQL에 등록된 사용자만 접속  
접속하는 서버 등록 시 여러 개 등록 가능  
Connection Name: 접속하는 이름

다른 서버로 접속하려면 [Local instance MySQL]에서 마우스 오른쪽을 클릭한 후 [Edit Connection]을 선택해서 추가로 접속할 서버를 등록하거나 편집

- [Connection] 탭  
   - Connection Method  
     Standard(TCP/IP ), Local Socket/Pipe, Standard TCP/IP over SSH 등 3가지 중에 선택 가능  
     대부분 Standard (TCP/IP) 사용

- [Parameters] 탭  
  - Hostname  
    - Localhost = 127.0.0.1 = 자신의 컴퓨터 = MySQL이 설치된 컴퓨터  
      접속할 컴퓨터가 외부에 있다면 접속할 서버 컴퓨터의 IP주소 입력  
    - Port   
      접속할 MySQL 포트 번호  
      특별한 경우가 아니면 3306  
    - Username, Password, 기본 DB 이름은 적절히 설정

- [SSL] 탭  
SSL (Secure Socket Layer)   
보안을 위한 암호 규약  
서버와 클라이언트가 통신할 때 암호화 통해 비밀 유지 + 보안 강화

- [Advanced] 탭  
프로토콜의 압축, 인증 방식 등을 설정

- [Remote Management] 탭  
  - 원격 관리 위해 설정하는 부분    
  - 이 부분이 활성화 되기 위해서는 Hostname 부분이 127.0.0.1을 제외한 실제 IP주소로 설정되어 있어야 함  
  - ‘Native Windows remote management’ 선택  
     MySQL 서버가 설치된 OS가 Windows인 경우에만 설정 가능  
  - ‘SSH login based management’는 SSH 서버 기반으로 원격 접속

- [System Profile] 탭  
  접속할 서버의 OS 종류 및 MySQL 설정 파일의 경로 등을 설정  
  - [Remote Management]에서 ‘Native Windows remote management’나 ‘SSH login based management’가 선택되어 있어야 활성화   
  - [System Type]은 FreeBSD, Linux, MacOS X, OpenSolaris, Windows 등 5가지 중 선택  
    [Installation Type]을 선택 가능   
  - [Configuration File] MySQL의 설정 파일이 경로와 함께 지정
  - [Configuration File Section] 서버의 서비스 이름 지정  
  - [MySQL Management] MySQL 서비스를 시작하거나 중지하는 시스템 명령어

## 5.3 사용자 관리  
---

현재까지 사용 방법은 MySQL 관리자인 root로 접속해 사용

실무에서는 MySQL 데이터베이스를 다양한 사용자나 응용프로그램에 접속해 사용

- 실습5  
사용자 및 역할/권한을 관리

**팀장님 (Director)**  
데이터베이스 관리자(DBA)의 역할 부여  

Workbench 실행하고 [Local instance MySQL]을 클릭해서 접속  
사용자를 생성하는 권한은 root에게만 있음  
[Navigator]의 [Management] 탭 [Users and Privileges] 클릭  
[Users and Privileges] 창에서 왼쪽 아래 [Add Account] 클릭한 후 [Login] 탭의 [Login Name]에 ‘director’ 입력  
비밀번호 입력하고 [Apply] 클릭 director 사용자 등록 확인  
[Account Limits] 탭으로 쿼리 한계 설정 0은 제한 없음   
[Administrative Roles] -  MySQL 자체에 대한 권한 설정

**사장님 (ceo))**  
MySQL의 모든 데이터에 읽기 권한 부여  

[Navigator]의 [Management] 탭 [Users and Privileges] 클릭  
[Users and Privileges] 창에서 왼쪽 아래 [Add Account] 클릭한 후 [Login] 탭의 [Login Name]에 ‘ceo’ 입력  
비밀번호 입력하고 [Apply] 클릭 ceo 사용자 등록 확인  
[Account Limits] 탭으로 쿼리 한계 설정 0은 제한 없음   
[Administrative Roles] -  MySQL 자체에 대한 권한 설정

[Global Privileges] 중에서 [SELECT] 체크
왼쪽 Role 중에 [Custom]이 자동으로 체크

**일반직원 (staff))**  
MySQL의 모든 데이터에 읽기,쓰기,스토어드 프로시저 등을 생성,수정 + employees 데이터베이스의 테이블에 대해서는 읽기만 권한 부여  


[Navigator]의 [Management] 탭 [Users and Privileges] 클릭  
[Users and Privileges] 창에서 왼쪽 아래 [Add Account] 클릭한 후 [Login] 탭의 [Login Name]에 ‘staff’ 입력  
[Schema Privileges] 탭 [Add Entry] 클릭  
[Select schema]에서 'shopdb'를 선택  
object rights에서는 [SELECT],[INSERT],[UPDATE],[DELETE]를 체크하고 DDL rights에서는 [CREATE ROUTINE],[AFTER ROUTINE]을 체크  
[Add Entry] employees에는 SELECT만 클릭
