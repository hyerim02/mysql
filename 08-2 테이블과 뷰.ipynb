## 열의 추가

회원 테이블에 회원의 홈페이지 주소를 추가하기

열을 추가하면 기본적으로 가장 끝에 추가됨  
열을 추가하면서 순서를 지정하려면 맨 뒤에 'FIRST' 나 'ALTER 열 이름'을 지정하면 됨  
'ALTER 열 이름'는 열 이름 다음 열에 생김

## 열의 삭제

제약 조건이 있는 경우 제약 조건을 삭제한 후에 열을 삭제해야 함
