## 8.1.4 임시 테이블

임시로 사용되는 테이블  
`CREATE TEMPORARY TABLE [IF NOT EXISTS] 테이블 이름  
(열 정의.......)`

TEMPORARY TABLE 써주는것 제외하고는 테이블 정의하는 방식과 동일

단 세션내에서만 존재하고 세션이 닫히면 삭제  
생성한 클라이언트에서만 접근 가능  
동일한 이름을 사용하면 임시 테이블이 삭제되기 전까지 임시 테이블만 사용 가능
