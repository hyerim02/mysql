## 4.1 프로젝트의 진행 단계

**프로젝트란**  
현실세계의 업무를 컴퓨터 시스템으로 옮겨놓는 일련의 과정  
대규모의 프로그램을 작성하기 위한 전체 과정  
초기에는 몇 명의 뛰어난 프로그래머에 의해 작성  
그런데 이제는 규모가 커지고, 사람들의 요구사항이 더욱 복잡해지면서 문제 발생

집 짓는 것과 비교할 수 있음  
옛날에 초가집이나 목조건물 지을때는 몇 명의 우수한 기술자로도 충분했으나 현재는 몇 명의 기술자로는 지을 수 없는 것과 같은 이치

계획 $\rightarrow$ 분석 $\rightarrow$ 설계  

분석과 설계 작업을 등한시하는 소프터웨어 분야의 고질적인 문제점때문에 이를 해결하기 위해 '소프트웨어 개발 방법론'이 나옴  
가장 오래되고 전통적으로 사용하는 모델이 **폭포수 모델**

프로젝트의 진행 단계가 명확해지는 장점이 있으나, 문제점 발생 시 다시 앞 단계로 거슬러 올라가기 어려운 단점이 있음.

## 4.2데이터베이스 모델링  
---

### 4.2.1 데이터베이스 모델링 개념  
---

**데이터베이스 모델링**  
현 세계에서 사용되는 작업이나 사물들을 DBMS의 데이터베이스 개체로 옮기기 위한 과정  
현실에서 쓰이는 것을 테이블로 변경하기 위한 자업

### 4.2.2 데이터베이스 모델링 실습  
---

- 개념적 모델링 : 업무 분석 단계  

- 논리적 모델링 : 업무 분석의 후반부와 시스템 설계의 전반부  

- 물리적 모델링 : 시스템 설계의 후반부

- 실습1  
새로운 쇼핑몰을 오픈  
매장을 찾는 고객의 명단을 기록하고 또 물건을 구매할 때 구매한 내역도 기록 $\rightarrow$  데이터베이스 모델링

- 실습2  
다이어그램으로 만들기

[File] $\rightarrow$ [New Model]  $\rightarrow$  [My SQL Model]이 열림  
마우스 오른쪽 버튼을 클릭한 후 [Edit Schema]를 선택하고 이름 바꾸고 닫음  
[Model Overview]의 [Add Diagram]을 더블클릭  
[place a New Table]아이콘을 클릭한 후에 빈 화면에 다시 마우스 클릭  
1:N관계를 맺어줌  
저장

모델링한 파일을 실제 데이터베이스에 적용  
[Open Model]을 열어서 [Database]에 [Forward Engineer]를 선택  
계속 Next누르고 finish

기존의 파일이 있는 경우 [Database]에 [Reverse Engineer]를 선택  
계속 Next누르고 finish
