## 6.1 SELECT문  

## SELECT...FROM

SELECT문은 가장 많이 사용  
처음에는 별거 아닌 것처럼 보이지만 점점 어려워짐  

### USE 

SELECT문을 사용하려면 먼저 사용할 데이터베이스를 정해야 함

**`USE ???;` $\rightarrow$ 지금부터 ???을 사용하겠으니, 모든 쿼리는 ???에서 수행해라**

또는 직접 선택해서 지정 가능

### SELECT & FROM

`SELECT * FROM titles;`  
* 은 모든 열을 의미  
FROM 다음에는 테이블/뷰

- 실습1  
데이터베이스 이름, 테이블 이름, 필드 이름이 정확히 기억나지 않거나 또는 각 이름의 철자가 확실하지 않을 때 조회하는 방법을 실습  
employees 데이터베이스에 있는 employees테이블의 first_name 및 gender열이라고 가정

1. 모든 창을 닫고 새 쿼리 창을 연다.  
2. 서버에 어떤 데이터베이스가 있는지 조회  
3. 데이터 베이스 지정  
4. 테이블 정보 조회  
5. 열 조회  
6. 데이터 조회

- 회원 가입을 하면, 기입한 회원 정보는 회원 테이블에 입력  
  아이디/이름/출생년도/거주지역/휴대폰 국번/휴대폰 번호/키/가입일  
- 물건을 구매하면, 구매한 정보는 구매 테이블에 기록  
- 배송 담당자가 구매 테이블을 통해서 주문한 물건을 준비하고, 구매 테이블의 아이디와 일치하는 회원 테이블의 아이디를 찾아서 배송

- 실습2  
앞으로 책의 전 과정에서 사용할 데이터베이스의 테이블을 생성  

1. DB만들기  
2. 테이블을 만들기  
3. 데이터 입력하기  
4. 데이터 확인하기  
5. 저장하기  
6. 초기화
