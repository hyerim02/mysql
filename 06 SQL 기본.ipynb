# 6.1 SELECT문  

## SELECT...FROM

SELECT문은 가장 많이 사용  
처음에는 별거 아닌 것처럼 보이지만 점점 어려워짐  

### USE 

SELECT문을 사용하려면 먼저 사용할 데이터베이스를 정해야 함

**`USE ???;` $\rightarrow$ 지금부터 ???을 사용하겠으니, 모든 쿼리는 ???에서 수행해라**

또는 직접 선택해서 지정 가능

### SELECT & FROM

`SELECT * FROM titles;`  
* 은 모든 열을 의미  
FROM 다음에는 테이블/뷰

- 실습1  
데이터베이스 이름, 테이블 이름, 필드 이름이 정확히 기억나지 않거나 또는 각 이름의 철자가 확실하지 않을 때 조회하는 방법을 실습  
employees 데이터베이스에 있는 employees테이블의 first_name 및 gender열이라고 가정

1. 모든 창을 닫고 새 쿼리 창을 연다.  
2. 서버에 어떤 데이터베이스가 있는지 조회  
3. 데이터 베이스 지정  
4. 테이블 정보 조회  
5. 열 조회  
6. 데이터 조회

- 회원 가입을 하면, 기입한 회원 정보는 회원 테이블에 입력  
  아이디/이름/출생년도/거주지역/휴대폰 국번/휴대폰 번호/키/가입일  
- 물건을 구매하면, 구매한 정보는 구매 테이블에 기록  
- 배송 담당자가 구매 테이블을 통해서 주문한 물건을 준비하고, 구매 테이블의 아이디와 일치하는 회원 테이블의 아이디를 찾아서 배송

- 실습2  
앞으로 책의 전 과정에서 사용할 데이터베이스의 테이블을 생성  

1. DB만들기  
2. 테이블을 만들기  
3. 데이터 입력하기  
4. 데이터 확인하기  
5. 저장하기  
6. 초기화

문제가 생겼을 때  
열려있는 쿼리 창을 닫고 파일을 다시 불러온다  
그리고나서 다시 실행  

## SELECT...FROM...WHERE

### WHERE절

조회하는 결과에 특정한 조건을 줘서 원하는 데이터만 보고 싶을 때 사용 

`SELECT * FROM 테이블 이름 WHERE 조건;`

### 연산자의 사용

1970년 이후에 출생하고 신장이 182 이상인 사람의 아이디와 이름을 조회

1970년 이후에 출생하거나 신장이 182 이상인 사람의 아이디와 이름을 조회

### BETWEEN...AND와 IN 그리고 LIKE

키가 180-183인 사람을 조회

동일한 방식으로 BETWEEN...AND를 사용가능

지역이 경남,전남,경북인 사람 정보를 확인

연속적인 값이 아닌 이산적인 값을 위해 IN을 사용할 수 있음

문자열의 내용을 검색하기 위해서는 LIKE 연산자를 사용할 수 있음  
성이 김씨이고 그 뒤는 무엇이든 허용  
맨 앞 글자가 한 글자이고, 그 다음이 종신인 사람을 조회

### ANY/ALL/SOME/서브쿼리

`서브쿼리`는 쿼리문 안애 또 쿼리문이 들어 있는 것

김경호보다 키가 크거나 같은 사람의 이름과 키를 출력하려면 WHERE 조건에 키를 직접 입력  
그런데 키를 직접 써주는 것이 아니라 쿼리를 통해서 하려고 함

지역이 경남인 사람의 키보다 크거나 같은 사람을 추출

173과 170 두가지 값이 반환되어서 에러

170 또는 173 이니까 170 이상인 사람들

170,173 둘다 만족해야하니까 173이상인 사람들

### ORDER BY 순서대로 정렬하여 출력

결과물에는 영향을 미치지 않지만 결과다 출력되는 순서를 조절하는 구문  
먼저 가입한 순서로 회원들을 출력

기본은 오름차순, 내림차순은 `DESC`

여러 개로 정렬  
키 $\rightarrow$ 이름
