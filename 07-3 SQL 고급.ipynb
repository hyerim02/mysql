## 스토어드 프로시저 만들고 사용하는 법

`DELIMITER $$
CREATE PROCEDURE 스토어드 프로시저이름()
BEGIN
....(SQL 프로그래밍 코딩)
END $$
DELIMITER;
CALL 스토어드 프로시저이름();`

`DELIMITER $$ ~ END $$`는 스토어드 프로시저의 코딩할 부분을 묶어줌  
`DELIMITER;`로 종료 문자를 세미콜론으로 바꿔야 함  
`CALL 스토어드 프로시저이름();` 호출
